```python
class selenium_scrape:
    def __init__(self,driver,url,name,price,year_of_purchase, km_driven,place, date_of_ad, owner, fuel):
        self.driver=driver
        self.url=url
        self.name=name
        self.price=price
        self.year_of_purchase=year_of_purchase
        self.km_driven=km_driven
        self.place=place
        self.date_of_ad=date_of_ad
        self.owner=owner
        self.fuel=fuel
    def connection(self):
        self.driver.maximize_window()
        time.sleep(3)
        return self.driver
    def web_search(self):
        web=self.driver.get(self.url)
        web1=self.driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div/div[1]/div[2]/div[1]/a')
        click=web1.click()#click on cars
        time.sleep(3)
 #       web2=self.driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/section/div/div/div[5]/div[1]/div/div[3]/div[4]/div[2]/div[2]/a[1]/div')
 #       click1=web2.click()
 #       time.sleep(3)
        web3=self.driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/section/div/div/div[5]/div[1]/div/div[3]/div[5]/div[2]/div[2]/div[1]/div')
        click2=web3.click()
        time.sleep(3)
        web4=self.driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/section/div/div/div[5]/div[1]/div/div[3]/div[8]/div[2]/div[2]/div[1]/div/input')
        click3=web4.click()
        time.sleep(3)
  #      web4=self.driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/section/div/div/div[5]/div[1]/div/div[3]/div[7]/div[2]/div[2]/a[1]/div/div[1]')
  #      click3=web4.click()
  #      time.sleep(3)
  #      web5=self.driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/section/div/div/div[5]/div[1]/div/div[3]/div[6]/div[2]/div[2]/div/div')
   #     click4=web5.click()
   #     time.sleep(3)
        return web, click,click2,click3
    def name_(self):
        for x in range(40,1000):
            try:
                btn=self.driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/section/div/div/div[5]/div[2]/div/div[3]/ul/li[{}]/div/button'.format(x))
                btn.click()
                time.sleep(3)
            except NoSuchElementException:
                continue
        new_web=self.driver.find_elements_by_xpath('//div[@class="_4aNdc"]')
        time.sleep(3)
        for x in new_web:
            self.name.append(x.text)
        return self.name
    def price_(self):
        new_web1=self.driver.find_elements_by_xpath('//span[@class="_3GOwr"]')
        time.sleep(3)
        for x in new_web1:
            self.price.append(x.text)
        return self.price
    def y_o_p(self):
        new_web2=self.driver.find_elements_by_xpath('//div[@class="KFHpP"]')
        time.sleep(3)
        yop=[]
        yp=[]
        for x in new_web2:
            yp.append(x.text)
        for x in yp:
            yop.append(x.split('-'))    
        for x in range(len(yop)):
            self.year_of_purchase.append(yop[x][0]) 
        for x in range(len(yop)):
            self.km_driven.append(yop) 
        return self.year_of_purchase, self.km_driven
    def place_(self):
        new_web3=self.driver.find_elements_by_xpath('//div[@class="_1zvfB"]')
        time.sleep(3)
        yop=[]
        yp=[]
        for x in new_web3:
            yp.append(x.text)
        for x in yp:
            yop.append(x.split('\n'))    
        for x in range(len(yop)):
            self.place.append(yop[x][0]) 
        for x in range(len(yop)):
            self.date_of_ad.append(yop[x][1]) 
        return self.place, self.date_of_ad
    def filters(self):
        for x in range(len(self.name)):
            self.owner.append('First')
            self.fuel.append('Petrol')
        return self.owner, self.fuel
    def data(self):
        data=pd.DataFrame({'Name':self.name, 'Price':self.price,'Year_Of_Purchase':self.year_of_purchase,
                           'km_driven':self.km_driven, 'Place':self.place, 'Date Of Ad':self.date_of_ad,
                           'Owner':self.owner, 'Fuel':self.fuel})
        return data
print('*'*30,'Establishing Connection','*'*30)
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
print('*'*30,'Connection Established Successfully','*'*30)
web='https://www.olx.in/'
name=[]
price=[]
year_of_purchase=[]
km_driven=[]
place=[]
date_of_ad=[]
owner=[]
fuel=[]
def main():
    sel=selenium_scrape(driver,web,name,price,year_of_purchase, km_driven,place, date_of_ad, owner, fuel)
    sel.connection()
    conn_sea=sel.web_search()
    names=sel.name_()
    prices=sel.price_()
    years=sel.y_o_p()
    places=sel.place_()
    static=sel.filters()
    data1=sel.data()
    print(names)
    print(len(names))
    print(prices)
    print(len(prices))
    print(years[0])
    print(len(years[0]))
    print(years[1])
    print(len(years[1]))
    print(places[0])
    print(len(places[0]))
    print(places[1])
    print(len(places[1]))
    print(static[0])
    print(len(static[0]))
    print(static[1])
    print(len(static[1]))
    print(data1)
    data1.to_csv('data_1.csv')

if __name__=='__main__':
    main()
```
